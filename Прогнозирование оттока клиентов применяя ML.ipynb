```python
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

–¢–æ –µ—Å—Ç—å, –µ—Å–ª–∏ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –≤ –æ—Ä–∏–≥–∏–Ω–∞–ª—å–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –±—ã–ª–æ 2:5, —Ç–æ –≤ –æ–±—É—á–∞—é—â–µ–π –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∞—Ö –æ–Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—Å—è —Ç–∞–∫–∏–º –∂–µ. –¢–∞–∫–æ–π –ø–æ–¥—Ö–æ–¥ –≤ —á–µ–º-—Ç–æ —Ä–µ—à–∞–µ—Ç –ø—Ä–æ–±–ª–µ–º—É –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤.

#### –û–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ train-–≤—ã–±–æ—Ä–∫–µ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π

```python
model = LogisticRegression(solver='liblinear')
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
probabilities =  model.predict_proba(X_test)[:,1]
```

```python
print('\tAccuracy: {:.2f}'.format(accuracy_score(y_test, predictions)))
print('\tPrecision: {:.2f}'.format(precision_score(y_test, predictions)))
print('\tRecall: {:.2f}'.format(recall_score(y_test, predictions)))
print('\tF1: {:.2f}'.format(f1_score(y_test, predictions)))
```

#### –û–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ train-–≤—ã–±–æ—Ä–∫–µ —Å–ª—É—á–∞–π–Ω—ã–º –ª–µ—Å–æ–º

```python
print('\tAccuracy: {:.2f}'.format(accuracy_score(y_test, rf_predictions)))
print('\tPrecision: {:.2f}'.format(precision_score(y_test, rf_predictions)))
print('\tRecall: {:.2f}'.format(recall_score(y_test, rf_predictions)))
print('\tF1: {:.2f}'.format(f1_score(y_test, rf_predictions)))
```
